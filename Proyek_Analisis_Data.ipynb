```python
datasets = [customers, items, payments, orders, products, category_translation]
titles = ["customers", "items", "payments", "orders", "products", "category_translation"]

#Membuat rangkuman seluruh dataset yang dipilih
#Menyiapkan tabel df kosong
info_df = pd.DataFrame({},)

#Menyiapkan kolom pertama df dengan judul dataset
info_df['dataset']= titles

#Memberikan keterangan nama kolom yang ada pada setiap dataset
info_df['cols'] = [', '.join([col for col, null in df.isnull().sum().items() ]) for df in datasets]

#Memberikan keterangan banyak baris yang ada pada dataset
info_df['rows_no']= [df.shape[0] for df in datasets]

#Memberikan keterangan banyak kolom yang ada pada dataset
info_df['cols_no']= [df.shape[1] for df in datasets]

#Memberikan keterangan banyak duplikat data pada dataset
info_df['dupl_no']= [df.duplicated().sum() for df in datasets]

#Memberikan keterangan banyak data yang null pada dataset
info_df['null_no']= [df.isnull().sum().sum() for df in datasets]

#Memberikan keterangan banyak kolom pada dataset yang terdapat data null
info_df['null_cols_no']= [len([col for col, null in df.isnull().sum().items() if null > 0]) for df in datasets]

#Memberikan keterangan pada kolom dataset mana saja yang terdapat data null
info_df['null_cols'] = [', '.join([col for col, null in df.isnull().sum().items() if null > 0]) for df in datasets]


info_df.style.background_gradient(cmap='Set2')
```

### Cleaning Data

```python
orders['order_purchase_timestamp']=pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at']=pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date']=pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date']=pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date']=pd.to_datetime(orders['order_estimated_delivery_date'])

orders.info()
```

```python
df = pd.merge(orders,customers)
df = df.merge(payments)
df = df.merge(items)
df = df.merge(products)
df = df.merge(category_translation)


df.dropna(inplace=True)  # Mempertahankan data-data yang valid tanpa Nan atau Null

df.info()
df.isnull().sum()  # Untuk mengetahui apakah masih ada missing value atau tidak
```

```python
df.to_csv('databersih.csv',index=False)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Produk apa saja yang paling banyak dibeli oleh pelanggan?

```python
top_10_cat = df["product_category_name_english"].value_counts().sort_values(ascending=False)[:10]
 
fig=plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
ax = sns.barplot(y=top_10_cat.index, x=top_10_cat.values, palette = 'Set2')
ax.bar_label(ax.containers[0])
plt.title('Top 10 Kategori Produk',fontsize=20)
```

### Pertanyaan 2: Bagaimana sebaran pelanggan e-Commerce tersebut?

```python
top_states = df["customer_state"].value_counts().sort_values(ascending=False)[:10]

fig=plt.figure(figsize=(10,5))
sns.set_style("darkgrid")
sns.barplot(y=top_states.values , x=top_states.index)
plt.title('Top 10 Kota Pelanggan',fontsize=20)
```

### Pertanyaan 3: Metode pembayaran apa saja yang dapat digunakan?

```python
df["payment_type"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    fontsize=15,
    figsize=(8,8))

plt.title('Tipe Pembayaran'.title() , fontsize=20);
```
